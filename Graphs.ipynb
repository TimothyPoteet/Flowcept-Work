```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('LLM Evaluations - Swallow Only Evaluations.csv')
```

# SCATTER PLOT OF DURATION vs WORD COUNT WITH BEST FIT

```python
# PLOT
plt.figure(figsize=(8, 6))
plt.scatter(df['Duration(s)'], df['Word_Count'], alpha=0.6, label='Data Points', color='blue')

# LINE OF BEST FIT
x = df['Duration(s)']
y = df['Word_Count']
slope, intercept = np.polyfit(x, y, 1)
line = slope * x + intercept
plt.plot(x, line, color='green', label=f'Best Fit Line: y = {slope:.2f}x + {intercept:.2f}')

# AVG WORD COUNT LINE
avg_word_count = y.mean()
plt.axhline(y=avg_word_count, color='red', linestyle='--', label=f'Avg Word Count = {avg_word_count:.1f}')

# lABELING
plt.xlabel('Duration (seconds)')
plt.ylabel('Word Count')
plt.title('Duration vs. Word Count with Best Fit Line')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# BAR CHART OF ACCURACY SCORE

```python
# TOTAL NUMBER OF 1'S 2'S & 3'S
score_counts = df['Accuracy_Score'].value_counts().sort_index()

# BAR CHART
plt.figure(figsize=(6, 4))
plt.bar(score_counts.index.astype(str), score_counts.values, color='orange')

# LABELS
plt.xlabel('Accuracy Score')
plt.ylabel('Count')
plt.title('Frequency of Accuracy Scores (1, 2, 3)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
